```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
```

### Параметры

```python
file_path = "student_scores.csv"
x_col = "Hours"
y_col = "Scores"
```

### Чтение файла

```python
data = pd.read_csv(file_path)
```

### Вывод статистики (количество, min, max, среднее и т.д.)

```python
print(data.describe())
```

### Вывод изображения исходных точек

```python
def plotData(data, x_col, y_col):
    plt.scatter(data[x_col], data[y_col], color='red', label='Исходные точки')
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    plt.title('Исходные точки')
    plt.legend()
    plt.grid()
    plt.show()

plotData(data, x_col, y_col)
```

### Реализация метода наименьших квадратов

```python
def linearRegression(data, x_col, y_col):
    X = data[x_col].values
    Y = data[y_col].values
    n = len(X)

    X_mean = sum(X) / n
    Y_mean = sum(Y) / n

    fraq1 = sum((X[i] * Y_mean  - X[i]*Y[i]) for i in range(n)) # числитель
    fraq2 = sum((X[i] * X_mean  - X[i]*X[i]) for i in range(n)) # знаменатель

    w1 = fraq1 / fraq2;
    w0 = Y_mean - w1 * X_mean

    return w0, w1
```

### Отрисуем регрессионную прямую

```python
params = linearRegression(data, x_col, y_col)
print(f"Параметры регрессионной прямой w0: {params[0]} w1: {params[1]}")

def plotWithRegressionLine(data, x_col, y_col, regressionLineParams):
    X = data[x_col].values
    b, k  = regressionLineParams;
    plt.scatter(data[x_col], data[y_col], color='red', label='Исходные точки')
    plt.plot(X, k * X + b, color='blue', label='Регрессионная прямая')
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    plt.title('Исходные точки + регрессиона прямая')
    plt.legend()
    plt.grid()
    plt.show()

plotWithRegressionLine(data, x_col, y_col, params)
```

### Отрисовка и штриховка квадратов ошибок

```python
def plotWithErrSquares(data, x_col, y_col, regressionLineParams):
    X = data[x_col].values
    Y = data[y_col].values
    b,k = regressionLineParams
    
    plt.scatter(X, Y, color='red', label='Исходные точки')
    plt.plot(X, k * X + b, color='blue', label='Регрессионная прямая')
    


    # Заштриховываем квадраты ошибок

    Y_pred = k*X + b    
    for i in range(len(X)):
        plt.plot([X[i], X[i]], [Y[i], Y_pred[i]], color='gray', linestyle='--')  # Вертикальные линии

    


    
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    plt.title('Квадраты ошибок')
    plt.legend()
    plt.show()


plotWithErrSquares(data, x_col, y_col, params)
```
