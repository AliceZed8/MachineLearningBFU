# Лабораторная работа 3.2 часть 2

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_iris, make_classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
%matplotlib inline
```

```python
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target
df.head()
```

Оставим только два признака для визуализации: petal length и petal width

```python
X = df[['petal length (cm)', 'petal width (cm)']].values
y = df['target'].values
```

Обучим модель многоклассовой логистической регрессии

```python
clf = LogisticRegression(random_state=0)
clf.fit(X, y)
```

### Визуализация

```python
# Построение сетки для 2 признаков
x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 300),
                     np.linspace(y_min, y_max, 300))

# Предсказания для каждой точки на сетке с использованием только двух признаков
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)


plt.figure(figsize=(10, 6))
plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.Paired)
scatter = plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired, edgecolors='k')
plt.xlabel('Petal length (cm)')
plt.ylabel('Petal width (cm)')
plt.title('Многоклассовая логистическая регрессия на Iris (2 признака)')
plt.legend(handles=scatter.legend_elements()[0], labels=iris.target_names.tolist())
plt.grid(True)
plt.show()
```
