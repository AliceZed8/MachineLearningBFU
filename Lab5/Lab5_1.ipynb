```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.metrics import classification_report, accuracy_score, roc_curve, roc_auc_score, precision_recall_curve, auc
from graphviz import Source
%matplotlib inline
```

```python
diabetes = pd.read_csv("diabetes.csv")
diabetes.head()
```

## Задание 1

```python
X = diabetes.drop(columns=['Outcome'])
y = diabetes['Outcome']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
logreg = LogisticRegression(max_iter=10000)
logreg.fit(X_train, y_train)
y_pred_logreg = logreg.predict(X_test)


dtree = DecisionTreeClassifier()
dtree.fit(X_train, y_train)
y_pred_dtree = dtree.predict(X_test)


print("Логистическая регрессия:")
print(classification_report(y_test, y_pred_logreg))
print("Accuracy:", accuracy_score(y_test, y_pred_logreg))
print("ROC AUC:", roc_auc_score(y_test, y_pred_logreg))


print("\nРешающее дерево:")
print(classification_report(y_test, y_pred_dtree))
print("Accuracy:", accuracy_score(y_test, y_pred_dtree))
print("ROC AUC:", roc_auc_score(y_test, y_pred_dtree))
```

Вывод:
- Логистическая регрессия продемонстрировала более высокую точность и сбалансированное разделение классов. Особенно хорошее соотношение precision и recall для класса 0 (здоровые), что делает эту модель более подходящей для данной задачи.
- Решающее дерево имеет более низкую производительность по метрикам и может страдать от переобучения, что снижает его способность обобщать на новых данных.

Для данной задачи классификации, логистическая регрессия выглядит более подходящей моделью, так как она дает лучшее разделение классов и имеет более высокую общую точность, особенно для сложных классов.

## Задание 2

Выбираем ROC AUC как метрику, так как она позволяет оценить качество модели независимо от соотношения классов и является хорошим показателем для задачи классификации с несбалансированными классами

```python
max_depth_values = np.arange(1, 21)
roc_auc_scores = []


for max_depth in max_depth_values:
    dt = DecisionTreeClassifier(max_depth=max_depth, random_state=42)
    dt.fit(X_train, y_train)
    y_pred_prob = dt.predict_proba(X_test)[:, 1]
    roc_auc_scores.append(roc_auc_score(y_test, y_pred_prob))


plt.figure(figsize=(8, 6))
plt.plot(max_depth_values, roc_auc_scores, marker='o', color='b')
plt.title('Зависимость ROC AUC от глубины дерева')
plt.xlabel('Глубина дерева')
plt.ylabel('ROC AUC')
plt.grid(True)
plt.show()
```

Оптимальная глубина дерева: Вероятно, наилучшая производительность достигается при глубине дерева от 2 до 5, где дерево еще не слишком сложное и может эффективно обобщать данные.

Переобучение: При глубине более 7 дерево слишком сильно подстраивается под обучающие данные, что приводит к ухудшению модели на новых данных (переобучение).

### Дерево

```python
dt = DecisionTreeClassifier(max_depth=5, random_state=42)
dt.fit(X_train, y_train)


dot_data = export_graphviz(dt, out_file=None, 
                           feature_names=X_train.columns,  
                           class_names=["0", "1"],  
                           filled=True, rounded=True,  
                           special_characters=True)  
graph = Source(dot_data)  
graph
```

### Важность признаков

```python
importances = dt.feature_importances_
indices = np.argsort(importances)[::-1]


plt.figure(figsize=(8, 5))
plt.title("Важность признаков")
plt.barh(range(X_train.shape[1]), importances[indices], align="center")
plt.yticks(range(X_train.shape[1]), X_train.columns[indices])
plt.xlabel("Важность")
plt.show()
```

### Кривые PR и ROC

```python
y_pred_prob = dt.predict_proba(X_test)[:, 1]

# ROC кривая
fpr, tpr, _ = roc_curve(y_test, y_pred_prob)
roc_auc = auc(fpr, tpr)

# PR кривая
precision, recall, _ = precision_recall_curve(y_test, y_pred_prob)
pr_auc = auc(recall, precision)

# Строим графики
plt.figure(figsize=(12, 6))

# ROC кривая
plt.subplot(1, 2, 1)
plt.plot(fpr, tpr, color="blue", lw=2, label=f"ROC AUC {roc_auc:.2f}")
plt.plot([0, 1], [0, 1], color="gray", linestyle="--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Кривая")
plt.legend(loc="lower right")

# PR кривая
plt.subplot(1, 2, 2)
plt.plot(recall, precision, color="blue", lw=2, label=f"PR AUC {pr_auc:.2f}")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("PR Кривая")
plt.legend(loc="lower left")

plt.tight_layout()
plt.show()
```
